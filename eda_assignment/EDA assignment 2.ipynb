7. What are the most popular product reviews?

```python
df.dtypes
```

```python
df['review_title'] = df['review_title'].fillna('')
df['review_content'] = df['review_content'].fillna('')

# Split the review_title and review_content columns into lists
df['review_title_list'] = df['review_title'].astype(str).str.split(',')
df['review_content_list'] = df['review_content'].astype(str).str.split(',')

# Ensure both lists have the same length by padding the shorter list with empty strings
max_len = df[['review_title_list', 'review_content_list']].applymap(len).max(axis=1)
df['review_title_list'] = df['review_title_list'].apply(lambda x: x + [''] * (max_len.max() - len(x)))
df['review_content_list'] = df['review_content_list'].apply(lambda x: x + [''] * (max_len.max() - len(x)))

# Explode the review_title_list and review_content_list columns to get individual reviews
df_exploded = df.explode(['review_title_list', 'review_content_list'])
```

```python
df_sorted = df_exploded.sort_values('rating_count', ascending=False).reset_index(drop=True)

# Select the top 10 most popular reviews
top_10_popular_reviews = df_sorted.head(10)

# Display the top 10 most popular reviews
print(top_10_popular_reviews[['product_name', 'review_title_list', 'review_content_list', 'rating_count']])
```

6. Among the top 5 most popular artists, which artist has the highest popularity on average?
 Calculate and display the average popularity for each artist

```python
import pandas as pd

artist_popularity = data.groupby('Artist')['Popularity'].sum()

# Get the top 5 most popular artists based on total popularity
top_5_artists = artist_popularity.nlargest(5)

# Calculate the average popularity for each of the top 5 artists
average_popularity = {}
for artist in top_5_artists.index:
    artist_tracks = data[data['Artist'] == artist]
    average_popularity[artist] = artist_tracks['Popularity'].mean()

# Determine which artist has the highest average popularity
highest_avg_popularity_artist = max(average_popularity, key=average_popularity.get)
highest_avg_popularity = average_popularity[highest_avg_popularity_artist]

# Display the average popularity for each artist
print("Average Popularity for Each Artist:")
for artist, avg_popularity in average_popularity.items():
    print(f"{artist}: {avg_popularity}")

# Display the artist with the highest average popularity among the top 5 most popular artists
print(f"\nThe artist with the highest average popularity among the top 5 most popular artists is '{highest_avg_popularity_artist}' with an average popularity of {highest_avg_popularity}.")
```
